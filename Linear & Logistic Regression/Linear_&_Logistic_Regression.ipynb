```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, r2_score, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Salary_Data.csv')
df
```

```python
x=df.drop(columns=['Salary'])
y=df['Salary']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
train_test_split(x,y,test_size=0.2)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
print(x_test)
print(y_test)
```

```python
y_predict=model.predict(x_test)
```

```python
y_predict
```

```python

mse=mean_squared_error(y_test,y_predict)
print(mse)
```

```python
sqrt_model=np.sqrt(mse)
print(sqrt_model)
```

```python
r2_model=r2_score(y_test,y_predict)
print(r2_model)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_predict,color='red')
plt.show()
```

```python
y_predict_train=model.predict(x_train)
plt.scatter(x_train,y_train)
plt.plot(x_train,y_predict_train,color='yellow')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

# LOGISTIC REGRESSION

### NOTE: When we have less independent columns or 1, the best classification model is logistical Model

```python
df=pd.read_csv('insurance_data.csv')
df
```

```python

print(df.shape)
df.info
```

```python
plt.scatter(df.age,df.bought_insurance,color='red',marker='o')
plt.title('Bought Insurance v/s Age')
plt.ylabel('Bought_Insurance')
plt.xlabel('Age')
plt.show()
```

```python
df.isnull().sum()
```

```python
X=df.drop(columns=['bought_insurance'])
y=df.bought_insurance
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
train_test_split(X,y,test_size=0.2)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
y_predict
```

```python
model.predict_proba(X_test)
```

```python
df_test=pd.DataFrame({'y_test':y_test,'y_predict':y_predict})
df_test
```
